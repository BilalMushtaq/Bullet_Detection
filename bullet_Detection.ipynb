```python
from ultralytics import YOLO
```

```python
import cv2, keyboard, time
import xlwings as xw
```

```python
row = 2
```

```python
workbook = xw.Book('records.xlsx')
worksheet = workbook.sheets('Sheet1')
worksheet.range('A1').value = 'Coordinates'
```

```python
model = YOLO('bullet_detect.pt')
#image = cv2.imread('target.png')
```

```python
def font_scale(text, width):
    for scale in reversed(range(0, 60, 1)):
        textSize = cv2.getTextSize(text, fontFace=cv2.FONT_HERSHEY_DUPLEX, fontScale=scale/10, thickness=1)
        new_width = textSize[0][0]
        if (new_width <= width):
            print(new_width)
            return scale/10
    return 1

def transformed_coor(x,y, image):
    new_x = x-(image.shape[1]//2)
    new_y = (image.shape[0]//2)-y
    
    return new_x, new_y
```

```python
cap = cv2.VideoCapture('test_video.mp4')

#cap = cv2.VideoCapture('trash/VID-20231010-WA0001.mp4')

while True:
    
    coor_list = []
    state, image = cap.read()
    
    if state==False:
        break
    
    res = model(image, conf=0.19)
    
    
    h, w = image.shape[:2]
    w = w//2
    h = h//2
    cv2.circle(image, (w,h), 5, (255,0,0), -1)
    cv2.line(image,(w,0), (w,h*2), (255,0,0),2)
    cv2.line(image,(0,h), (w*2,h), (255,0,0),2)
    for labels in range(len(res[0].boxes.data)):
        
        
        x1 = int(res[0].boxes.data[labels][0])
        y1 = int(res[0].boxes.data[labels][1])
        x2 = int(res[0].boxes.data[labels][2])
        y2 = int(res[0].boxes.data[labels][3])
        xc = (x1+x2)//2
        yc = (y1+y2)//2
        
        #cv2.circle(image, (xc,yc) ,5,(0,0,255),-1)
        coor = str(transformed_coor(xc,yc,image))
        cv2.putText(image, (coor), (xc,yc-1), cv2.FONT_HERSHEY_SIMPLEX ,
                    0.5, (0, 0, 255) , 1, cv2.LINE_AA)
        coor_list.append(coor)
    
    if keyboard.is_pressed('enter'):
        worksheet.range(f'A{row}').value = f'{coor_list}, num bullets:{len(coor_list)}'
        row+=1
        time.sleep(0.5)
        
    cv2.imshow('f',image)
    c = cv2.waitKey(1)
    if c==27:
        break
cv2.destroyAllWindows()
cap.release()
```
