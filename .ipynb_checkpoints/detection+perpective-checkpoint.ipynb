```python
from ultralytics import YOLO
import cv2, numpy
```

```python
model = YOLO('bullet_detect.pt')
```

```python
source_coordinates=[]
target_coordinates = numpy.float32([(0,0), (0,400), (400,0), (400,400)])
loop_numbers=1

coordinates=[]
coordinate_flag = False

max_distance = 0
```

```python
def click_event(event, x, y, flags, params): 
    global loop_numbers
    # checking for left mouse clicks 
    if event == cv2.EVENT_LBUTTONDOWN:
        if loop_numbers<5:
            source_coordinates.append((x,y))
            
            loop_numbers+=1
        cv2.circle(frame, (x,y), 5, (0,0,255), -1 )
        cv2.imshow('frame', frame)
```

```python
cap = cv2.VideoCapture('test_video.mp4')
```

```python
# font 
font = cv2.FONT_HERSHEY_SIMPLEX 
  
# org 
org = (50, 50) 
  
# fontScale 
fontScale = 0.5
   
# Blue color in BGR 
color = (255, 0, 0) 
  
# Line thickness of 2 px 
thickness = 1

ep=False
```

```python
while True:
    state, frame = cap.read()
    
    if state==False:
        break
    frame = cv2.resize(frame, (600,600))
    cv2.imshow('frame', frame)
    
    cv2.setMouseCallback('frame', click_event)
    
    if loop_numbers>=5:
        
        source_coordinates = numpy.float32(source_coordinates)
        #transformed_image = frame[tl[0]:tr[0], tl[1]:bl[1]]
        trans_matrix = cv2.getPerspectiveTransform(source_coordinates, target_coordinates)
        transformed_image = cv2.warpPerspective(frame, trans_matrix, (400,400))

        res = model(transformed_image, conf=0.19)

        for label_idx, label_data in enumerate(res[0].boxes.xywh):
            if len(coordinates)==0:
                coordinates.append((int(label_data[0]), int(label_data[1])))
            for idx, data in enumerate(coordinates):
                lx = label_data[0]
                ly = label_data[1]
                dx = data[0]
                dy = data[1]
                
                if lx >= dx+5 or lx <= dx-5 or ly >= dy+5 or ly <=dy-5:
                    coordinate_flag = True
                else:
                    coordinate_flag = False
                    break
            if coordinate_flag == True:
                coordinates.append((int(label_data[0]), int(label_data[1])))
                
        
        for idx, data in enumerate(coordinates):
            for idx2, data2 in enumerate(coordinates):
                if data!=data2:
                    distance = numpy.sqrt(((data2[0]-data[0])**2)+((data2[1]-data[1])**2))
                    
                    if distance > max_distance:
                        ep=True
                        extreme_points = (idx, idx2)
                        max_distance = distance
                
            difference = numpy. sqrt((data))
            cv2.putText(transformed_image, str(idx), (data),  font,  
                   fontScale, color, thickness, cv2.LINE_AA)
    
        cv2.putText(transformed_image, str("%.2f" %(((max_distance)/100)*12))+' in' , (180,350), font, 1, color, 2, cv2.LINE_AA)
        if ep==True:
            cv2.line(transformed_image, coordinates[extreme_points[0]] , coordinates[extreme_points[1]], (0,0,255) , 2)
        
        cv2.imshow('transformed image', transformed_image)
                
            
            
    
    c = cv2.waitKey(1)
    if c==27:
        break
        
cv2.destroyAllWindows()
cap.release()
```
